```python
import instaloader
import matplotlib.pyplot as plt

username = input("Enter the Instagram username you want to analyze: ")

bot = instaloader.Instaloader()
profile = instaloader.Profile.from_username(bot.context, username)

print("Username: ", profile.username)
print("User ID: ", profile.userid)
print("Number of Posts: ", profile.mediacount)
print("Followers Count: ", profile.followers)
print("Following Count: ", profile.followees)
print("Bio: ", profile.biography)
print("External URL: ", profile.external_url)
print("------------------------------")

posts = profile.get_posts()

def display_post_data(posts):
    for index, post in enumerate(posts, 1):
        print(f"Post {index}:")
        print("Caption:", post.caption)
        print("Likes:", post.likes)
        print("Comments:", post.comments)
        print("Date:", post.date)

        sizes = [post.likes, post.comments, post.likes + post.comments]
        labels = ['Likes', 'Comments', 'Shares']  # Assuming shares = likes + comments for simplicity

        fig, ax = plt.subplots()
        ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'lightcoral'])
        ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
        ax.set_title(f'Social Engagement for Post {index}')

        # Draw a circle at the center to make it a donut chart
        centre_circle = plt.Circle((0, 0), 0.70, fc='white')
        fig.gca().add_artist(centre_circle)

        plt.show()

display_post_data(posts)
```
